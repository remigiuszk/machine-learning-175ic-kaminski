#importy


```
import sklearn as sk
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

#deklaracja danych

```
data={
    'size':['L','M','S', 'XL', 'S', 'M', 'M', 'XL', 'L', 'M', 'L', 'S', 'S', 'M', 'XL', 'L', 'S', 'M', 'S', 'L'],
    'color':['green','yellow','yellow','black','blue','blue','orange','yellow','purple','black','grey','brown','green', 'red', 'yellow','red','black','brown','blue','white' ],
    'gender':['female','female','female','male','female','male','male','female','male','male','male','female','male','female','female','female','female','male', 'male', 'male'],
    'price':[149.0,79.0    ,111.0 ,45.0 ,21.0 ,90.0 ,180.0 ,210.0 ,333.0 ,69.0 ,85.0 ,165.0 ,120.0 ,199.0 ,231.0 ,200.0 ,399.0 ,23.0 ,16.0 ,90.0],
    'weight':[380, 800, 230, 500, 550, 440, 609, 423, 831, 351, 388, 490, 540, 390, 470, 450, 400, 500, 346, 210],
    'bought':['no','yes','yes','yes','no','no','yes','yes','no','no','yes','no','no','no','yes','yes','no','no','yes','no'] 
}
```

#dataframe z biblioteki pandas

```
df_raw = pd.DataFrame(data=data)
df = df_raw.copy()

for col in ['size','color','gender','bought']:
  df[col] = df[col].astype('category')

for col in['weight']:
  df[col] = df[col].astype('float')

df.info()

df.describe(include=['category']).T
df
```

#przekształcanie danych typu string

```
le = LabelEncoder()
le.fit_transform(df['bought'])
le.classes_
df['bought'] = le.fit_transform(df['bought'])
df
```

#kategoryzowanie danych wejśćiowych

```
 df = pd.get_dummies(data=df, drop_first=True)
 df
```

#standaryzacja wartości

```

scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
```

#porównanie tabel

```
print(df)
print(df_raw)
```
