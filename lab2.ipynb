```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
data={
    'size':['L','M','S', 'XL', 'S', 'M', 'M', 'XL', 'L', 'M', 'L', 'S', 'S', 'M', 'XL', 'L', 'S', 'M', 'S', np.nan],
    'color':['green','yellow',np.nan,'black','blue','blue','orange','yellow','purple','black','grey','brown',np.nan, 'red', np.nan,'red','black','brown','blue','white' ],
    'gender':[np.nan,'female',np.nan,'male','female','male','male','female','male','male',np.nan,'female','male','female','female',np.nan,'female','male', 'male', 'male'],
    'price':[149.0,79.0,111.0 ,45.0 ,21.0 ,90.0 ,180.0 ,210.0 ,np.nan ,69.0 ,85.0 ,165.0 ,120.0 ,np.nan ,231.0 ,200.0 ,399.0 ,23.0 ,np.nan ,90.0],
    'weight':[380, 800, 230, 500, 550, np.nan, 609, 423, 831, 351, 388, 490, 540, 390, 470, 450, 400, 500, 346, 210],
    'bought':['no','yes','yes','yes','no','no','yes','yes','no','yes','yes','no','no','no','yes','yes','no','no','yes','no'] 
}
df_raw = pd.DataFrame(data=data)
df = df_raw.copy()
df


```

# 1. pomiń dane, które mają 3 lub więcej pustych wartości 

```
df.dropna(thresh = (len(df.keys())-2), inplace=True)
df
```

# 2. wypełnij puste miejsca w kolumnie price wartością średnią  

```
df['price'].fillna(df['price'].mean(), inplace=True)
df
```

# 3. wypełnij puste miejsca w kolumnie weight za pomocą interpolacji 

```
df['weight'].fillna(df['weight'].interpolate(), inplace=True)
df
```

# 4. wypełnij puste miejsca w kolumnie size, color i gender wartościami stałymi (np. unisex, black, male) 

```
df['size'].fillna('unisize', inplace=True)
df['color'].fillna('black', inplace=True)
df['gender'].fillna('other', inplace=True)
df
```

# 4. dodaj do słownika kolumnę cost, która reprezentuje cenę jako jedną z klas: cheap, normal_price, expensive; podziału ceny na te klasy dokonaj automatycznie (argument bins metody pandas.cut) 

```
df['cost'] = pd.cut(df['price'], bins = 3, labels=('cheap', 'normal_price', 'expensive'))
df
```

# 6. dostosuj dane do modelu tak, by przedstawiały wartość binarną (np. bought: [‘yes’, ‘no’, ‘yes’] na bought_yes: [1,0,1]) 

```
df['bought_y'] = df['bought'].apply(lambda x: 1 if 'yes' in x else 0)
df
```

# 7. podziel słownik na dane wejściowe i wartość wyjściową, a następnie podziel na zbiór uczący i testujący 

```
df1=df[['price','cost']]
df2=df['bought_y']

X_train, X_test, y_train, y_test = train_test_split(df1, df2, train_size=0.2)
```
